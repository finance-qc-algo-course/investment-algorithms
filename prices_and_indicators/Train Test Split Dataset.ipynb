```python
import numpy as np
import pandas as pd
import os
```

```python
DATA_PATH = "./data/prices/tech_companies_15minute.csv"
TRAIN_PATH = "./data/prices/tech_companies_15minute_train.csv"
TEST_PATH = "./data/prices/tech_companies_15minute_test.csv"

TEST_RATIO = 0.2
```

```python
data = pd.read_csv(DATA_PATH, index_col=0)
data.sample(10)
```

```python
dates = data.apply(lambda row: pd.Timestamp(row["date"]), axis=1)
dates
```

```python
from_date = dates.min()
to_date = dates.max()

time_border = to_date - (to_date - from_date) * TEST_RATIO
train_data = data[dates <= time_border]
test_data = data[dates > time_border]

assert train_data.shape[0] + test_data.shape[0] == data.shape[0]

train_data.sample(10)
```

```python
train_data.to_csv(TRAIN_PATH)
test_data.to_csv(TEST_PATH)
```
